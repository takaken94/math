```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def neuron(x, w, b):
    return sigmoid(w * x -b)
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
# 重み 1 バイアス 0 赤のグラフ
y = neuron(x, 1, 0)
plt.plot(x, y, color="r") 

# 重み 0.5 バイアス 0 青のグラフ
y = neuron(x, 0.5, 0)
plt.plot(x, y, color="b")

# 重み 2 バイアス 0 緑のグラフ
y = neuron(x, 2, 0)
plt.plot(x, y, color="g") 
plt.ylim(0, 1.5) 
plt.show()
```
