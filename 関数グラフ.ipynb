```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# 関数 f(x) = 1/3*x**3 - x + 1 を定義
def f(x):
    return (1/3) * x**3 - x + 1

# x の範囲を設定
x = np.linspace(-3, 3, 400)

# y の値を計算
y = f(x)

# グラフを作成
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y, label='f(x) = 1/3*x**3 - x + 1', color='b')
ax.set_title('f(x) = 1/3*x**3 - x + 1 のグラフ')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(-3, 3)  # x軸の表示範囲
ax.set_ylim(-6, 6)  # y軸の表示範囲
ax.set_xticks(np.arange(-3, 4, 1))  # x軸の目盛りを1毎に設定
ax.set_yticks(np.arange(-6, 7, 1))  # y軸の目盛りを1毎に設定
ax.grid(True)
ax.legend()

# X軸とY軸を太い罫線に設定
ax.axhline(y=0, color='k', linewidth=2.5)
ax.axvline(x=0, color='k', linewidth=2.5)

plt.show()
```
