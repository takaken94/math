```python
x = np.arange(-3, 3, 0.01)
y = 2*x**2
```

```python
plt.xlim(-3, 3) # x軸の表示範囲
plt.ylim(0, 15) # y軸の表示範囲
plt.title("2次関数グラフ",fontsize=10) # (5)タイトル
plt.xlabel("x", fontsize=10) # x軸のラベル
plt.ylabel("y", fontsize=10) # y軸のラベル
plt.grid(True) # 目盛線の表示
plt.plot(x, y)
```
