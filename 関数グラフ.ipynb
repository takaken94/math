```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# 関数 f(x) = 1/3*x**3 - x + 1 を定義
def f(x):
    return (1/3) * x**3 - x + 1

# x の範囲を設定
x = np.linspace(-3.5, 3.5, 400)

# y の値を計算
y = f(x)

# グラフを作成
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y, label='f(x) = 1/3*x**3 - x + 1', color='b')
ax.set_title('f(x) = 1/3*x**3 - x + 1 のグラフ')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(-3.5, 3.5)  # x軸の表示範囲
ax.set_ylim(-7.5, 7.5)  # y軸の表示範囲
ax.set_xticks(np.arange(-3, 4, 1))  # x軸の目盛りを1毎に設定
ax.set_yticks(np.arange(-6, 7, 1))  # y軸の目盛りを1毎に設定
ax.grid(True)
ax.legend()

# X軸とY軸を太い罫線に設定
ax.axhline(y=0, color='k', linewidth=2.5)
ax.axvline(x=0, color='k', linewidth=2.5)

plt.show()
```

```python
# 関数 y = x ** 4 と y = x ** 3 を定義
def f1(x):
    return x ** 4

def f2(x):
    return x ** 3

# x の範囲を設定
x = np.linspace(-3, 3, 400)

# y の値を計算
y1 = f1(x)
y2 = f2(x)

# グラフを作成
plt.figure(figsize=(10, 10))
plt.plot(x, y1, label='y = x ** 4', color='b')
plt.plot(x, y2, label='y = x ** 3', color='r')
plt.title('y = x ** 4 と y = x ** 3 のグラフ')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5, 5)  # x軸の表示範囲
#plt.ylim(-30, 30)  # y軸の表示範囲
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# xの範囲（0は除く、正の値のみ）
x = np.linspace(0.01, 5, 400)

# 各関数の定義
y1 = 2 * np.sqrt(x)
y2 = 1 / x
y3 = 2 * np.sqrt(x) + 1 / x

# y3の極小値を求める
# y3 = 2√x + 1/x の微分 = 1/√x - 1/x^2
# これが0になるxを求める
# 1/√x = 1/x^2 → x^2 = √x → x^4 = x^0.5 → x^3.5 = 1 → x = 1^(1/3.5) = 1
xmin = 1
ymin = 2 * np.sqrt(xmin) + 1 / xmin

plt.figure(figsize=(10, 8))
plt.plot(x, y1, label='y = 2√x', color='b')
plt.plot(x, y2, label='y = 1/x', color='r')
plt.plot(x, y3, label='y = 2√x + 1/x', color='g')
plt.scatter(xmin, ymin, color='k', s=80, label=f'極小値 ({xmin}, {ymin:.2f})', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 2√x, y = 1/x, y = 2√x + 1/x のグラフ')
plt.legend()
plt.grid(True)
plt.ylim(-10, 10)
plt.xlim(0, 5)
plt.show()
```
