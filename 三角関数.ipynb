# 三角関数

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

## cos, sin グラフ

```python
x = np.linspace(-0.2, 2*np.pi + 0.2, 100)
y1 = np.cos(x)
y2 = np.sin(x)

plt.title("y = cos, y = sin グラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y1, color="r", label="cos(x)")
plt.plot(x, y2, color="b", label="sin(x)")
plt.plot(0, 1, marker='.', color='r')
plt.plot(np.pi / 2, 0, marker='.', color='r')
plt.plot(0, 0, marker='.', color='b')
plt.plot(np.pi / 2, 1, marker='.', color='b')
plt.legend(loc="lower left")
plt.grid(ls=':')
plt.show()
```

## tan グラフ

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.tan(x)

plt.title("y = tan のグラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y, color="g", label="tan(x)")
plt.legend(loc="lower left")
plt.grid(ls=':')
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.tan(x)

threshold = 10
y[y >  threshold] = np.inf
y[y < -threshold] = np.inf

fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(x, y, color="g", label="tan(x)")
ax.legend(loc="lower left")
ax.grid(ls=':')
ax.set_ylim([-4, 4])
ax.set_title("y = tan のグラフ")
plt.show()
```

## sin(x) + 2sin(2x)

```python
x = np.linspace(-0.2, 2 * np.pi + 0.2, 100)
y1 = np.sin(x)
y2 = 2*np.sin(2*x)
y3 = np.sin(x) + 2*np.sin(2*x)

plt.plot(x, y1, color="b", label="sin(x)")
plt.plot(x, y2, color="g", label="2sin(2x)")
plt.plot(x, y3, color="r", label="sin(x) + 2sin(2x)")
plt.legend(loc="lower left")
plt.grid(ls=':')
plt.show()
```

## 単位円を描く

```python
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.grid(ls=':')
plt.show()
```
